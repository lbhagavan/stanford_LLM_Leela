# Prompt engineering - few shot - Classification


````python
def chat(message):
    """
    Send a message to the OpenAI GPT-3.5 model and return its response.

    This function interacts with the OpenAI API, specifically using the GPT-3.5-turbo model. It takes a user's message as input, sends it to the model, and returns the model's text-only response. The function ensures the AI's output is concise by providing a system-level instruction.

    Parameters:
    message (str): A string containing the user's message to the AI.

    Returns:
    str: The text response generated by the GPT-3.5 model.
    """

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        temperature=0.1,
        # response_format={ "type": "json_object" },
        messages=[
            {"role": "system", "content": """ Few-shot classifications. After studying these examples, please classify the new text message at the end.

              Example 1:

              Alert: " You must review and closely monitor the bubble housing prices? It's extremely emergency for decision-making, you must review and report the housing price areas with its Peers."
              Classification: Extreme Emergency

              Example 2:

              Alert: "Please review the data and deliver the summary report within 2 hours! We need it ready by upcoming meeting with CEO by 2.00 pm today!"
              Classification: Urgent

              Example 3:

              Alert: "Let's plan to have lunch together for next 2 week to discuss the project. Take your times. Upon your availablty. When possible"
              Classification: Low Priority

              Example 4:

              Alert: "ALERT!!!: The Stock market is crashed by 25% due to scammer. Urgency!"
              Classification: Extreme Emergency

              Example 5:

              Alert: There are so many miss spellings and typo errors in the report. Please correct it asap.
              Classification: Low
              """},
            {"role": "user", "content": f"Classify the following message: {message}"}
        ]
    )

    text_only = response.choices[0].message.content
    return text_only
````

````python
def chat(message):
    """
    Send a message to the OpenAI GPT-4o model and return its response.

    This function interacts with the OpenAI API, specifically using the GPT-4o model. It takes a user's message as input, sends it to the model, and returns the model's text-only response. The function ensures the AI's output is concise by providing a system-level instruction.

    Parameters:
    message (str): A string containing the user's message to the AI.

    Returns:
    str: The text response generated by the GPT model.
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": """
            You are a super Genius AI Physician nominated for the 2024 Nobel Prize in Physics. You must provide precise and accurate answers in plain text, without using LaTeX formatting.

            Below are examples of questions and how to calculate the answer:

              Example 1: Arithmetic Problem
              Prompt: "If Giant Tech Stock's price now is trading at $30 per stock and a credit rating agency announces the upgraded company rating to the 'positive outlook' affecting its stock trading price suddenly increases by 7.5% after the news, how much does the Giant Tech Company's stock trading price after the upgraded company outlook?"
              Chain of Thought Answer:
                Calculate the latest stock's price gaining from the positive outlook: 7.5% increases from $30 is $30 * 1.075 = $32.25.
                The stock price increases by $2.25 after the upgraded positive outlook news.

              Example 2: Logic Puzzle
              Prompt: "Given that A, B, C, and D can be only positive numbers, if A is greater than B, and B is greater than C, and D is greater than C, but less than B. If A is equal to 5, what are the possible positive numbers that D and C can be?"
              Chain of Thought Answer:
                You start with A > B and A = 5, then B can be either 4, 3, 2, or 1.
                Then, C < D and D < B. You may say that if B = 4, 3, or 2 respectively, then D can be either 3, 2, or 1.
                However, C < D, meaning that C is the lesser positive number among A, B, and C. Thus, D cannot be 1 because C cannot be 0 either. You may say that D can be either 3 or 2.
                Therefore, A = 5, B = 4, D can be either 3 or 2, and C can be either 2 or 1.

              Example 3: BMI Calculation
              Prompt: "Calculate and classify BMI for Pen's weight of 60 kg and her height of 156 cm."
              Chain of Thought Answer:
                Convert height from centimeters to meters: 156 cm is 1.56 m.
                Calculate BMI using the formula: weight / (height in meters)^2.
                For 60 kg and 1.56 m: 60 / (1.56^2) = 24.65.
                Therefore, Pen's BMI is 24.65.
                Classify BMI based on the WHO's standard:
                - Less than 16: Severely Underweight
                - 16 to 16.99: Moderately Underweight
                - 17 to 18.49: Mildly Underweight
                - 18.5 to 24.99: Normal Weight
                - 25 to 29.99: Overweight (Pre-Obesity)
                - 30 to 34.99: Obesity Class I
                - 35 to 39.99: Obesity Class II
                - 40 and above: Obesity Class III (Severe Obesity)
                Therefore, Pen's BMI is 24.65, which is classified as Normal Weight.

              """},
            {"role": "user", "content": f"Answer the following question: {message}"}
        ]
    )
    text_only = response.choices[0].message.content
    return text_only
````

````python
from langchain.chains.combine_documents.stuff import StuffDocumentsChain
from langchain.chains.llm import LLMChain
from langchain.prompts import PromptTemplate

# Define prompt
prompt_template = """Write an executive summary of the top-5 highest cities' price, show the price number, and reasons in a maximum of 5 bullets of the following text enclosed within three backticks:
```{text}```
EXECUTIVE SUMMARY:"""
prompt = PromptTemplate.from_template(prompt_template)

# Define LLM chain
llm = ChatOpenAI(temperature=0, model_name="gpt-4o", api_key=open_ai_key)
llm_chain = LLMChain(llm=llm, prompt=prompt)

# Define StuffDocumentsChain
stuff_chain = StuffDocumentsChain(llm_chain=llm_chain, document_variable_name="text")

res = stuff_chain.invoke(pages[0:6])
````

#Extra Credit:  Exploring more funs on 1. using langchain to create my personal assistant and 2. using langchain to creat storytelling ✌

````python
# Step 1: Install necessary libraries
!pip install openai langchain

# Step 2: Set up the environment and import necessary modules
import os
from langchain.chains.llm import LLMChain
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI

# Manually set the environment variable for testing purposes
os.environ['open_api_key'] = 'open_ai_key'  # Replace with your actual OpenAI API key

# Retrieve the OpenAI API key from the notebook environment variable
openai_api_key = os.getenv('open_api_key')
if not openai_api_key:
    raise ValueError("OpenAI API key not found in environment variables")

# Define the LLM
llm = ChatOpenAI(api_key=openai_api_key, model="gpt-4o")

# Step 3: Personal Assistant
# Define the personal information
personal_info = """
My Hobbies: listen to the music, play tennis, talk and Fight with ChatGPT
My favorite song: I believe I can fly
My Favorite sport: Tennis
My challenge : Play snowboarding with toe and hill position from the top of the Niseko Mountain in Hokkaido
My favorite color : blue
My favorite Place: Chicago
"""

# Define the prompt template for the personal assistant
assistant_prompt = PromptTemplate(
    input_variables=["task"],
    template="""
You are my personal assistant. Here is some information about me:
{personal_info}

Your task is to: {task}
"""
)

# Create the chain
assistant_chain = LLMChain(llm=llm, prompt=assistant_prompt)

# Function to perform a task
def perform_task(task):
    response = assistant_chain.run({"personal_info": personal_info, "task": task})
    return response
````

````python
# Test the personal assistant
print(perform_task("Set a gentle reminder to drink and listen to the music tomorrow at 6 PM after finishing the TECH16 HW."))
print(perform_task("Recommend some good places to visit in Chicago."))
````

#Extra Credit --Create Storytelling using langchain

````python
# Step 1: Install necessary libraries
!pip install openai langchain

# Step 2: Set up the environment and import necessary modules
import os
from langchain.chains.llm import LLMChain
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI

# Manually set the environment variable for testing purposes
os.environ['open_api_key'] = 'open_ai_key'  # Replace with your actual OpenAI API key

# Retrieve the OpenAI API key from the notebook environment variable
openai_api_key = os.getenv('open_api_key')
if not openai_api_key:
    raise ValueError("OpenAI API key not found in environment variables")

# Define the LLM
llm = ChatOpenAI(api_key=openai_api_key, model="gpt-4o")

# Step 4: Interactive Storytelling
# Define the prompt template for storytelling
story_prompt = PromptTemplate(
    input_variables=["story"],
    template="{story}\nContinue the story from here:"
)

# Create the chain
story_chain = LLMChain(llm=llm, prompt=story_prompt)

# Function to continue the story
def continue_story(current_story):
    response = story_chain.run({"story": current_story})
    return response
````

````python
# Test the interactive storytelling
initial_story = """
My Japanese Shiba Inu name is Soba meaning the Japanese Street food. She is female shiba but does not look like female dog. She is naughty, crazy, and hungry dog. Her tail color is weird with mix brown and black color. She is the last dog sold out with huge discount from the Shiba Inu farm in Rayong Province. She is stubborn and does not like by barking me when I tell her to stop and point my finger to her
"""
print(continue_story(initial_story))
````
