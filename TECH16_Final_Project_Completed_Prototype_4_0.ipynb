```
import requests
from bs4 import BeautifulSoup

# Function to fetch and print tickers from a given URL and aria-label
def fetch_tickers(url, aria_label):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    tickers = []
    for row in soup.find_all('tr'):
        symbol_cell = row.find('td', {'aria-label': aria_label})
        if symbol_cell:
            tickers.append(symbol_cell.text)
    return tickers

# URLs and aria-labels for different categories
urls = {
    "Global Equities": ("https://finance.yahoo.com/world-indices/", 'Symbol'),
    "US Treasury Rates": ("https://finance.yahoo.com/bonds/", 'Symbol'),
    "Cryptocurrencies": ("https://finance.yahoo.com/crypto/", 'Symbol'),
    "Currencies": ("https://finance.yahoo.com/currencies/", 'Symbol')

}

# Fetch and print tickers for each category
for category, (url, aria_label) in urls.items():
    tickers = fetch_tickers(url, aria_label)
    print(f"{category} Tickers:")
    for ticker in tickers:
        print(ticker)
    print()  # Print a newline for better separation between categories
```

```
def chat(message):
    """
    Send a message to the OpenAI GPT-4o model and return its response.

    This function interacts with the OpenAI API, specifically using the GPT-4o model. It takes a user's message as input, sends it to the model, and returns the model's text-only response. The function ensures the AI's output is concise by providing a system-level instruction.

    Parameters:
    message (str): A string containing the user's message to the AI.

    Returns:
    str: The text response generated by the GPT model.
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system", "content": """
            You are a super Genius AI Physician nominated for the 2024 Nobel Prize in Physics. You must provide precise and accurate answers in plain text, without using LaTeX formatting.

            Below are examples of questions and how to calculate the answer:

              Example 1: Arithmetic Problem
              Prompt: "If Giant Tech Stock's price now is trading at $30 per stock and a credit rating agency announces the upgraded company rating to the 'positive outlook' affecting its stock trading price suddenly increases by 7.5% after the news, how much does the Giant Tech Company's stock trading price after the upgraded company outlook?"
              Chain of Thought Answer:
                Calculate the latest stock's price gaining from the positive outlook: 7.5% increases from $30 is $30 * 1.075 = $32.25.
                The stock price increases by $2.25 after the upgraded positive outlook news.

              Example 2: Logic Puzzle
              Prompt: "Given that A, B, C, and D can be only positive numbers, if A is greater than B, and B is greater than C, and D is greater than C, but less than B. If A is equal to 5, what are the possible positive numbers that D and C can be?"
              Chain of Thought Answer:
                You start with A > B and A = 5, then B can be either 4, 3, 2, or 1.
                Then, C < D and D < B. You may say that if B = 4, 3, or 2 respectively, then D can be either 3, 2, or 1.
                However, C < D, meaning that C is the lesser positive number among A, B, and C. Thus, D cannot be 1 because C cannot be 0 either. You may say that D can be either 3 or 2.
                Therefore, A = 5, B = 4, D can be either 3 or 2, and C can be either 2 or 1.

              Example 3: BMI Calculation
              Prompt: "Calculate and classify BMI for Pen's weight of 60 kg and her height of 156 cm."
              Chain of Thought Answer:
                Convert height from centimeters to meters: 156 cm is 1.56 m.
                Calculate BMI using the formula: weight / (height in meters)^2.
                For 60 kg and 1.56 m: 60 / (1.56^2) = 24.65.
                Therefore, Pen's BMI is 24.65.
                Classify BMI based on the WHO's standard:
                - Less than 16: Severely Underweight
                - 16 to 16.99: Moderately Underweight
                - 17 to 18.49: Mildly Underweight
                - 18.5 to 24.99: Normal Weight
                - 25 to 29.99: Overweight (Pre-Obesity)
                - 30 to 34.99: Obesity Class I
                - 35 to 39.99: Obesity Class II
                - 40 and above: Obesity Class III (Severe Obesity)
                Therefore, Pen's BMI is 24.65, which is classified as Normal Weight.

              """},
            {"role": "user", "content": f"Answer the following question: {message}"}
        ]
    )
    text_only = response.choices[0].message.content
    return text_only
```

```
# Load the BTC data into a DataFrame
btc_df = pd.read_csv('BTC Daily.csv')
btc_df['Date'] = pd.to_datetime(btc_df['Date'])
btc_df.set_index('Date', inplace=True)

# Load the FX data into a DataFrame
fx_df = pd.read_csv('USD_Index.csv')
fx_df['Date'] = pd.to_datetime(fx_df['Date'])
fx_df.set_index('Date', inplace=True)

# Calculate daily returns for BTC
btc_df['Daily_Return'] = btc_df['BTC_Close'].pct_change()

# Calculate daily returns for FX
fx_df['Daily_Return'] = fx_df['USD_Index'].pct_change()

# Calculate annual returns for BTC from 2011 to 2024
btc_df['Year'] = btc_df.index.year
annual_BTC_returns = btc_df[btc_df['Year'].between(2011, 2024)].groupby('Year')['Daily_Return'].apply(lambda x: (1 + x).prod() - 1)

# Calculate annual returns for FX from 2011 to 2024
fx_df['Year'] = fx_df.index.year
annual_FX_returns = fx_df[fx_df['Year'].between(2011, 2024)].groupby('Year')['Daily_Return'].apply(lambda x: (1 + x).prod() - 1)

# Display the annual returns for BTC
annual_BTC_returns_df = annual_BTC_returns.reset_index()
annual_BTC_returns_df.columns = ['Year', 'Annual_Return']
print("Annual BTC Returns:")
print(annual_BTC_returns_df)

# Display the annual returns for FX
annual_FX_returns_df = annual_FX_returns.reset_index()
annual_FX_returns_df.columns = ['Year', 'Annual_Return']
print("Annual FX Returns:")
print(annual_FX_returns_df)

# Display the annual returns for BTC in percentage
annual_BTC_returns_df['Annual_Return'] = annual_BTC_returns_df['Annual_Return'] * 100
print("Annual BTC Returns in Percentage:")
print(annual_BTC_returns_df)

# Display the annual returns for FX in percentage
annual_FX_returns_df['Annual_Return'] = annual_FX_returns_df['Annual_Return'] * 100
print("Annual FX Returns in Percentage:")
print(annual_FX_returns_df)


# Save the annual returns to CSV files
annual_BTC_returns_df.to_csv("annual_BTC_returns.csv", index=False)
annual_FX_returns_df.to_csv("annual_FX_returns.csv", index=False)
```

```
# Function to expand keywords with synonyms and antonyms
def expand_keywords(keywords, synonym_dict, antonym_dict):
    expanded_keywords = set(keywords)  # Use a set to avoid duplicates
    for keyword in keywords:
        for key, synonyms in synonym_dict.items():
            if key in keyword:
                for synonym in synonyms:
                    expanded_keywords.add(keyword.replace(key, synonym))
        for key, antonyms in antonym_dict.items():
            if key in keyword:
                for antonym in antonyms:
                    expanded_keywords.add(keyword.replace(key, antonym))
    return list(expanded_keywords)
```

```
def replace_synonyms(text, synonym_dict):
    """Replaces words in a text with their synonyms from a dictionary."""
    for word, synonyms in synonym_dict.items():
        for synonym in synonyms:
            text = text.replace(synonym, word)
    return text
```
